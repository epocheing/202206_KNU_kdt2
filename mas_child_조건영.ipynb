```python
# 필요한 열만 추출
df_2018_final=df_2018.iloc[:,[0,5,14,22]]
df_2018_final.rename(columns={'총계':'총 인원수','총계.1':'동/식물원 입장 수','총계.2':'테마가든 입장 수'},inplace=True)
df_2018_final
```

```python
# 2019년도 데이터 합치기
df_2019_1_4=pd.concat([df20191,df20192,df20193,df20194])
df_2019_5_12=pd.concat([df20195,df20196,df20197,df20198,df20199,df201910,df201911,df201912])
df_2019_5_12.columns=['날짜', '요일', '날씨', '유료합계', '무료합계', '총계', '외국인계', '유료소계', '어른', '청소년',
       '어린이', '외국인', '단체입장', '무료소계', '총계.1', '유료소계.1', '어른.1', '청소년.1',
       '어린이.1', '외국인.1', '단체입장.1', '무료소계.1', '총계.2']
df_2019=pd.concat([df_2019_1_4,df_2019_5_12])
df_2019
```

```python
# 19년도 데이터에서 필요한 열만 추출
df_2019_final=df_2019.iloc[:,[0,5,14,22]]
df_2019_final.rename(columns={'총계':'총 인원수','총계.1':'동/식물원 입장 수','총계.2':'테마가든 입장 수'},inplace=True)
df_2019_final
```

```python
# 18년도 19년도 데이터 합치기
df_1819=pd.concat([df_2018_final,df_2019_final])
df_1819
```

```python
# 데이터 확인
df_1819.info()
```

```python
# 타입바꾸기
df_1819['날짜']=pd.to_datetime(df_1819['날짜'])
df_1819.info()
```

```python
# 각 날짜의 입장객 수의 순위
df_1819.sort_values('총 인원수',ascending=False)
```

```python
# 어린이날과 크리스마스 입장객수 알아보기
chi_mas_su=df_1819[(df_1819['날짜']=='2018-05-05') | (df_1819['날짜']=='2018-12-25') | (df_1819['날짜']=='2019-05-05') | (df_1819['날짜']=='2019-12-25')]
chi_mas_su
```

```python
chi_mas_su['동/식물원 입장 비율']=chi_mas_su['동/식물원 입장 수']/chi_mas_su['총 인원수']*100
chi_mas_su['테마가든 입장 비율']=chi_mas_su['테마가든 입장 수']/chi_mas_su['총 인원수']*100
chi_mas_su
```

```python
# 서울대공원의 2019년 일일 입장객 수 그래프
df_1819.plot('날짜',['동/식물원 입장 수','테마가든 입장 수'],kind='line',figsize=(20,8))
plt.show()

# 5월 5일이 가장 제일 위에 있고
# 12월 25일은 추운 겨울인 12월인 다른 날에 비해 높게 쏟아있음
```
