```python
import pandas as pd
FILE1='../week3_team5/Data_신나령/datatrend2021.xls'
FILE2='../week3_team5/Data_신나령/2020naver_datalab_shoppingInsight_category_일간_data_20220711174443.csv'
FILE3='../week3_team5/Data_신나령/2019naver_datalab_shoppingInsight_category_일간_data_20220711174040.csv'
FILE4='../week3_team5/Data_신나령/2018_naver_datalab_shoppingInsight_category_일간_data_20220711172039.csv'
FILE5='../week3_team5/Data_신나령/relatedQueries.csv'
FILE6='../week3_team5/Data_신나령/2021빼빼로,가래떡.xls'
pd.set_option('display.max_rows',410)
df1=pd.read_excel(FILE1)
df2=pd.read_csv(FILE2,sep='\t')
df3=pd.read_csv(FILE3,sep='\t')
df4=pd.read_csv(FILE4,sep='\t')
df5=pd.read_csv(FILE5,sep='\t')
df6=pd.read_excel(FILE6)
```

### (1) 2021년 검색 데이터

```python
df1.head(10)
```

```python
# 2021년 필요없는 행 삭제
df1.drop([0,1,2,3,4,5],inplace=True)
```

```python
# 결측치 확인
df1.isnull().sum()
```

```python
# 결측치 있는 행 없애고 다시 확인
df1.dropna(inplace=True)
df1.isnull().sum()
```

```python
df1.head()
```

```python
def date1():
    dd1=df1['2021-01-01 ~ 2021-12-31']=='통합검색, 성별 검색량 비교'
    date1=[]
    for i in range(1,len(dd1)):
        if dd1.iloc[i]==True: break
        else: date1.append(df1.iloc[i])
    dateDF1=pd.DataFrame(date1)
    dateDF1=dateDF1.rename(columns={'기간':'date','2021-01-01 ~ 2021-12-31':'빼빼로'}) # 컬럼명 식별 편하게 바꿔 줌
    dateDF1.set_index('date',inplace=True) # 날짜로 인덱스 설정
    return dateDF1
date1()
```

### (10) 2021년 전체 빼빼로 & 가래떡 검색량 비교

```python
# 2021년 빼빼로, 가래떡 비교 필요없는 행 삭제
df6.drop([0,1,2,3,4,5],inplace=True)
```

```python
df6.dropna(inplace=True)
df6.isnull().sum()
```

```python
def date2():
    dd2=df6['Unnamed: 2']=='가래떡'
    date2=[]
    for i in range(1,len(dd2)):
        if dd2.iloc[i]==True: break
        else: date2.append(df6.iloc[i])
    dateDF2=pd.DataFrame(date2)
    dateDF2=dateDF2.rename(columns={'기간':'date','2021-01-01 ~ 2021-12-31':'빼빼로','Unnamed: 2':'가래떡'}) # 컬럼명 식별 편하게 바꿔 줌
    dateDF2.set_index('date',inplace=True) # 날짜로 인덱스 설정
    return dateDF2
date2()
```

```python
# 2021년 '빼빼로 & 가래떡' 검색량 그래프
plt.rcParams["figure.figsize"]=(15,7)#그래프 사이즈 설정
date2().plot(color=['green','orange'])
plt.title('<2021년 빼빼로 & 가래떡 검색량 비교>',size=13)
plt.xlabel('<Month>',size=13)#X축 이름
plt.ylabel('<Pepero Search>',size=13)#Y축 이름
plt.show()
```
