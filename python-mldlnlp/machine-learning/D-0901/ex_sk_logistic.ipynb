## Logistic Regression
---
- 인공신경망의 기본 개념과 동일
- Regression이지만 이진분류로 사용
- 선형모델 기반으로 y = wx+b 그래도 사용
- y = wx+b의 결과 -> 분류 변환 함수 sigmoid()

### 성적 - 합격/불합격 분류 실습

```python
# 사용 모듈 로딩
from sklearn.linear_model import LogisticRegression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### [1] 데이터 준비

```python
x = [78, 82, 91, 94, 97]    # 점수
y = [0, 0, 1, 1, 1]         # 합격여부 ( 합격 = 1, 불합격 = 0)
```

```python
# 학습에 사용할 수 있도록 형변환
arr_x = np.array(x).reshape(-1, 1)
arr_y = np.array(y).reshape(-1, 1)
```

```python
# 데이터 체크
print(f"arr_x = {arr_x.shape}, arr_y = {arr_y.shape}")
```

#### [2] 데이터 확인 -> 시각화

```python
# 현제 데이터 시각화
plt.scatter(x, y)
plt.xlabel("Score")
plt.ylabel("Result")
plt.show()
```

### [3] 모델 생성 & 학습

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(arr_x, y)
```

```python
# 학습 후 설정된 속성
print(f"분류 종류 : {model.classes_}")
print(f"가중치 값 : {model.n_features_in_}")
print(f"기 울 기 : {model.coef_}")
print(f"절     편 : {model.intercept_}")
```

### [4] 평가 -> score() 학습용 데이터 점수, 테스트용 데이터 점수

```python
print(f"model.score = {model.score(arr_x, y)}")
```

### [5] 테스트 -> predict(): 새로운 데이터로 진행 

```python
pred_value = np.array([60, 70, 80, 90, 95, 100]).reshape(-1, 1)

print(f"predict vale: {pred_value}")
print(f"predict: {model.predict(pred_value)}")
print(f"predict_proba: {model.predict_proba(pred_value)}")
```

```python
# 결정함수
model.decision_function([[82]])
```

```python
# y = Wx + b : 모델 파라미터
result = (model.coef_ * 82) + model.intercept_
print('result : ', result)
```
