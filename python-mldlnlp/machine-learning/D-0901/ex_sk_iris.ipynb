## iris dataset 활용 분류

```python
# 모듈 로딩
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### [1] 데이터 로딩 및 확인

```python
data, target = load_iris(as_frame = True, return_X_y = True)
```

```python
data.head(3)
```

```python
data.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
```

### [2] 학습 데이터 준비

```python
# 데이터 -> petal_length, petal_width
# 타 겟 -> 0, 2

data = data.iloc[:,[2,3]]
```

```python
target = pd.DataFrame(target)
```

```python
a = pd.concat([data, target], axis = 1)
```

```python
a.drop(a.loc[a["target"] == 1].index, inplace = True)
```

```python
a.reset_index(drop = True, inplace = True)
```

```python
data = a.iloc[:, [0,1]]
```

```python
target = a["target"]
```

### [3] 모델 분류

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, random_state=5, test_size = 0.2, stratify= target
)
```

```python
train_input.shape
```

```python
test_input.shape
```

### [4] 모델 학습

```python
logi = LogisticRegression()

logi.fit(train_input, train_target)
```

```python
train_score = logi.score(train_input, train_target)
test_score = logi.score(test_input, test_target)

print("train score: ", train_score)
print("test score: ", test_score)
```
