## SVM(Support Vector Machine)

- ML에서 오랫동안 연구되었고 성능이 좋은 모델
- DL보다 성능이 좋은 경우도 있음

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

### [1] 데이터 준비

```python
digit = load_digits()

digit.keys()
```

```python
# data, target 분리
data = digit['data']
target = digit['target']

print(f"data : {data.shape}, target : {target.shape}")
```

```python
print(f"Target : {target[1]}, images[1] \n\n {digit['images'][1]}")
```

```python
np.unique(target)
```

```python
plt.figure(figsize=(10,7))

for i in range(0,10):
    plt.gray()
    plt.subplot(2, 5, i+1)
    plt.imshow(digit['images'][i])

plt.show()
```

### [2] 데이터 나누기

```python
train_input, test_input, train_target, test_target = train_test_split(
    data, target, random_state = 42
)
```

### [3] 모델 학습

```python
# SVC
svc = SVC()

svc.fit(train_input, train_target)
print("train score: " ,svc.score(train_input, train_target))
print("test score: ", svc.score(test_input, test_target))
```

```python
# SVC linear
svc_linear = SVC(
    kernel = "linear",
)

svc_linear.fit(train_input, train_target)
print("train score: " ,svc_linear.score(train_input, train_target))
print("test score: ", svc_linear.score(test_input, test_target))
```
