# 퍼셉트론 (Perceptron)
---
- 뉴런을 컴퓨터 프로그램화 한 클래스
- 단순한 구조의 분류 모델

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import Perceptron
```

## [1] AND 분류
---
x1 x2 y  
0     0     0  
1     0     0  
0     1     0  
1     1     1  

```python
X1 = [[0., 0.], [1., 0.], [0., 1.], [1., 1.]]
y1 = [0, 0, 0, 1]
```

```python
and_model = Perceptron()
and_model.fit(X1, y1)
```

```python
# 모델 객체 속성 출력 함수
def show_attribute(model):
    print(f"classes_ : {model.classes_}")
    print(f"coef_ : {model.coef_}")
    print(f"intercept_ : {model.intercept_}")
    print(f"n_iter_ : {model.n_iter_}")
```

```python
show_attribute(and_model)
```

```python
# 예측
pre_and = and_model.predict(X1)
```

```python
print(f"정답 : {y1}")
print(f"예측 : {pre_and}")
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"{accuracy_score(y1, pre_and)}")
```

## [2] OR 연산 분류

```python
X2 = [[0., 0.], [1., 0.], [0., 1.], [1., 1.]]
y2 = [0, 1, 1, 1]
```

```python
or_model = Perceptron()
or_model.fit(X2, y2)
```

```python
show_attribute(or_model)
```

```python
# 예측
pre_or = or_model.predict(X2)
```

```python
print(f"정답 : {y2}")
print(f"예측 : {pre_or}")
```

```python
print(f"{accuracy_score(y2, pre_or)}")
```
