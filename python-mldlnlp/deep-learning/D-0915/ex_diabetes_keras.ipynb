```python
def show_val_count(df:pd.DataFrame):
    for col in df.columns:
        val_count = df[col].value_counts()
        print(f"{col}의 value_counts: ", val_count, sep="\n")
```

```python
def show_zero_val(df:pd.DataFrame):
    for col in df.columns:
        if list(df[col]).count(0) > 0:
            print(f"{col}의 0 갯수: {list(df[col]).count(0)}개")
```
