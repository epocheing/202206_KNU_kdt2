# 모델 속성 및 W, b 초기값 고장
---
- 모델 속성 확인
- 모델 생성 시 랜덤으로 초기화되는 W, b 값 고정

```python
from tensorflow import random
from keras import models, layers
import numpy as np
```

```python
# 데이터 생성
X = np.array([[-5, 2], [-10, -2], [9, 15], [4, 6], [-12, -2], [12, 15], [-7, -1]])
y = np.array([2,6,5,4,7,8,1])
```

```python
random.set_seed(72)
```

```python
# 모델 구상 및 생성
model = models.Sequential()

model.add(layers.Dense(15, input_shape=(2,), name="l1"))
model.add(layers.Dense(1, activation="linear"))
```

```python
model.summary()
```

```python
# 모델 학습
model.compile(
    optimizer="adam",
    loss="mse"
)

his = model.fit(
    X, y, epochs=10, verbose=0
)
```

```python
l1 = model.layers[0]
```
