<span style='color:Blue'> **인플레이션 현황 분석을 통한 투자 사업분야 선정** </span>
==============================================
**************
팀이름 : 굴려줄게 돈조  
팀원 : 송인욱, 신민수, 심정윤, 곽영효, 김수아  
--------------------------------------------------
#### 개요: 인플레이션이 경제에 미치는 영향과 밀접한 사업분야 현황을 비교하여 수혜를 보는 투자분야 탐색
    - 송인욱: 환률의 변동과 수출중심기업 분석
    - 곽영효: 식품 및 요식업계 현황분석
    - 신민수: 인플레이션 철강 분야에 미친 영향 및 한국 철강기업별 이익비교
    - 심정윤: 인플레이션으로 인한 목재 가격 변동 및 가구제조업 투자 여부 분석
    - 김수아: 인플레이션이 반도체 분야에 미친 영향 분석
