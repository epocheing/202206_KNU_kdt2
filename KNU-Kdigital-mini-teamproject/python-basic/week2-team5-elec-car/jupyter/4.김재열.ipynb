```python
import pandas as pd

DIR_PATH='../Data_jaeyeol/'
FILE_NAME=DIR_PATH+'수아2)한국전력공사_서울시 전기차 충전소 충전량_20220331.csv'
FILE_NAME
```

```python
# 윈도우에서 사용하는 한글 인코딩 CP949
elecDF=pd.read_csv(FILE_NAME,encoding='CP949')
elecDF
```

```python
# 데이터 분석에 필요한 열 추출
elecT=elecDF[['충전구분','충전소명','충전시작시각','충전종료시각']]
elecT
```

```python
# 데이터 타입이 object로 된 것을 확인
elecT.info()
```

```python
# 데이터 타입이 object로 된 것을 확인
elecT['충전시작시각']=pd.to_datetime(elecT['충전시작시각'])
elecT['충전종료시각']=pd.to_datetime(elecT['충전종료시각'])
elecT.info()
```

```python
# 결측치 확인
elecT.isnull().sum()
```

```python
# 결측치가 있는 행을 삭제
delnull=elecT.dropna()
delnull
```

```python
# 날짜 시간 계산을 위해 '충전시작시각'을 인덱스로 지정
TimeSet=delnull.set_index('충전시작시각')
TimeSet
```

```python
# datetime 함수에서 요일을 나타내는 dayofweek를 사용 (월 ~ 일 : 0 ~ 6)
TimeSet['dayofweek']=TimeSet['충전종료시각'].dt.dayofweek
TimeSet
```

```python
# 주말에 해당하는 5, 6값의 갯수 확인
weekend=TimeSet.dayofweek>4
weekend.value_counts()
```

```python
# 주말 데이터프레임 확인
WeekEndDF=TimeSet[weekend]
WeekEndDF
```

```python
# 평일에 해당하는 0, 1, 2, 3, 4 값의 갯수 확인
weekday=TimeSet.dayofweek<=4
weekday.value_counts()
```

```python
# 평일 데이터프레임 확인
WeekDayDF=TimeSet[weekday]
WeekDayDF
```

```python
# weekend=(TimeSet['dayofweek']==5) | (TimeSet['dayofweek']==6)
# weekend.value_counts()
```

```python
# 주말 시간대별 충전횟수를 카운드한후 시간대별로 정렬
WeekEndCount=WeekEndDF.index.hour.value_counts().sort_index()
WeekEndCount
```

```python
# 평일 시간대별 충전횟수를 카운드한후 시간대별로 정렬
WeekDayCount=WeekDayDF.index.hour.value_counts().sort_index()
WeekDayCount
```

```python
# 일주일 전체 시간대별 충전횟수를 시간대별로 정렬
WholeWeek=TimeSet.index.hour.value_counts().sort_index()
WholeWeek
```

```python
import matplotlib.pyplot as plt
```

```python
# Matplotlib 한글 Font 설정
from matplotlib import font_manager as fm, rc

# 한글 폰트 설정
font_path='../Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
fig, ax = plt.subplots(figsize=(13,10))
plt.plot(WholeWeek.index,WholeWeek.values, 'r--')     # 일주일 전체 시간대별 충전량
plt.plot(WeekDayCount.index,WeekDayCount.values, 'y--')    # 평일 전체 시간대별 충전량
plt.plot(WeekEndCount.index,WeekEndCount.values, 'b--')    # 주말 전체 시간대별 충전량
plt.ylabel('충전횟수')      # y 축 라벨
plt.xlabel('시간')
plt.title('일주일 시간대별 충전량')    # 그래프 타이틀
plt.legend(['일주일 전체','평일','주말'])              # 범례
plt.show()
```

```python
# 충전구분 열에서 '급속'을 원소로 가진 데이터프레임 
FastCharge=TimeSet['충전구분']=='급속'
FastChargeDF=TimeSet[FastCharge] 
FastChargeDF
```

```python
# 충전구분 열에서 '완속'을 원소로 가진 데이터프레임 
SlowCharge=TimeSet['충전구분']=='완속'
SlowChargeDF=TimeSet[SlowCharge]
SlowChargeDF
```

```python
# 고속충전 시간대별 충전횟수를 카운드한후 시간대별로 정렬
FastChargeDF=FastChargeDF.index.hour.value_counts().sort_index()
FastChargeDF
```

```python
# 완속충전 시간대별 충전횟수를 카운드한후 시간대별로 정렬
SlowChargeDF=SlowChargeDF.index.hour.value_counts().sort_index()
SlowChargeDF
```

```python
fig, ax = plt.subplots(figsize=(13,10))
plt.plot(FastChargeDF.index,FastChargeDF.values, 'r--')
plt.plot(SlowChargeDF.index,SlowChargeDF.values, 'b--')
plt.ylabel('충전횟수')      # y 축 라벨
plt.xlabel('시간')
plt.title('충전속도별 시간대 충전량')    # 그래프 타이틀
plt.legend(['고속충전', '완속충전'])              # 범례
ChargeSpeed=plt.show()
ChargeSpeed
```
