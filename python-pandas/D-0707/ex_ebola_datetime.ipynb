```python
import pandas as pd
from datetime import date

DIR = '../Data/'
FILE = 'country_timeseries.csv'
```

```python
ebolaDF = pd.read_csv(DIR+FILE)
ebolaDF.info()
```

```python
ebolaDF['Date'] = pd.to_datetime(ebolaDF['Date'])
```

```python
ebolaDF = ebolaDF.dropna(axis=0, how='all', thresh=2)
```

```python
ebolaDF = ebolaDF.fillna(0)
```

```python
ebolaDF.head()
```

```python
# ebola 바이러스 발병 후 몇일이 지났는지에 대한 컬럼 추가
ebolaDF['Date']
```

```python
today = date.today()
today = pd.to_datetime(today)
today
```

```python
try:
    test = today - ebolaDF['Date']
    ebolaDF.insert(2,'howDay',test)
    ebolaDF
except Exception:
    pass
type(test[0])
```
