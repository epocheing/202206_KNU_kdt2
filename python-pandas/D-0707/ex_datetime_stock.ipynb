# Datetime => Index로 설정 실습
- set_index(컬럼) : 인덱스

```python
import pandas as pd

DIR = '../Data/'
NAME = 'stock-data.csv'
```

```python
stockDF = pd.read_csv(DIR + NAME)
stockDF.info()
```

```python
stockDF.head(3)
```

```python
stockDF.isnull().sum()
```

```python
# 전처리 => NaN, Duplicate, DateType
# Date 컬럼 = Object => datetime64

# stockDF['Date'] = stockDF['Date'].astype('datetime64')

stockDF['Date'] = pd.to_datetime(stockDF['Date'])
stockDF
```

```python
stockDF[ stockDF['Date'].dt.month >= 7 ]
```

```python
result1 = stockDF['Date'].dt.month == 6
result2 = stockDF['Date'].dt.day >= 20
result3 = stockDF['Date'].dt.day < 26

stockDF[result1][result2][result3]
```

```python
try:
    stockDF.set_index('Date',inplace=True)
except Exception:
    pass
```
