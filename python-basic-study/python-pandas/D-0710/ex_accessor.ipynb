# 접근자(Accessor) 실습
- DF/SR에 속성으로 제공되며 다른 객체의 기능/속성 사용할 수 있도록 해줌
- 종류
    - Series.dt : datetime 관련 속성 사용 할 수 있도록 해줌
    - Series.str : string 관련 메서드 사용 할 수 있도록 해줌
    - DataFrame.spare : spare 관련 속성/메서드 사용할 수 있도록 해줌 

```python
# module
import pandas as pd, random

# 데이터생성
df = pd.DataFrame(
    {
        "name": [
            "Kim aa",
            "Park dd",
            "Song ii",
            "Oh great",
            "Lee sunshin",
            "Jay park",
            "Hong dangmu",
            "Show memoney",
            "Break World",
            "Yang achi",
        ],
        "birth": [i for i in pd.date_range('2008-01-10', periods=10, freq='M')],
        "age": [random.randint(15,35) for count in range(10)],
    }
)
```

## str 데이터로 구성된 컬럼의 str 접근자
- 전체 컬럼의 요소에 str관련 메서드 사용 가능하도록 하는 접근자
- 사용법
    - 객체변수[컬럼명].str.메서드() => upper, lower, split ...
    - 객체변수[컬럼명].str.get(인덱스) => 분리 문자리스트에서 요소 추출

```python
# 문자열 분리
flName = df['name'].str.split(' ')
type(flName)
```

```python
# 분리된 문자열 리스트에서 특정 값 가져오기
# df['last_name'] = flName.str.get(0)
# df['first_name'] = flName.str.get(1)
```

```python
value = df['name'].str.split(expand=True)
type(value)
```

```python
value.shape, value.ndim, value.columns
```

```python
df2 = pd.concat([df, value], axis=1)
df2 = df2.rename(columns={0: 'last_name', 1: 'first_name'})
```

## 실습 ----------------------------
- stock_data.csv 파일 불러와서 실습
- 날짜별로 분류하기

```python
stockDF = pd.read_csv('../Data/stock-data.csv')
stockDF.info()
```

```python
stockDF.Date.unique()
```

```python
date = stockDF['Date'].str.split('-', expand=True)
date = date.rename(columns={0: 'year', 1: 'month', 2: 'day'})
stockDF = pd.concat([stockDF, date], axis=1)
stockDF = stockDF.sort_values('day')
```
