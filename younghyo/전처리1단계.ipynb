```python
import pandas as pd
import os
```

```python
path = 'C:/RProject/team2_orange/younghyo/AmongUsData/'
files = os.listdir(path)
df = pd.concat([pd.read_csv(path + i) for i in files ])
df = df.reset_index(drop=True)
```

```python
df.drop(['Game Completed Date', 'Rank Change', 'Region/Game Code'], axis = 1, inplace = True)
```

```python
df.replace('-', '', inplace = True)
```

```python
int(df['Game Length'][0].split()[0][:2])*60 + int(df['Game Length'][0].split()[1][:2])
```

```python
for i in range(len(df.index)):
    df['Game Length'][i] = int(df['Game Length'][i].split()[0][:2])*60 + int(df['Game Length'][i].split()[1][:2])
```

```python
df.columns = ['Team', 'Outcome', 'Task_Completed', 'All_Tasks_Completed', 'Murdered',
       'Imposter_Kills', 'Game_Length', 'Ejected', 'Sabotages_Fixed',
       'Time_to_complete_all_tasks']
```
