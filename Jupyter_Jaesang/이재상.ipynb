```python
# 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Matplotlib 한글 Font 설정
from matplotlib import font_manager as fm, rc
from matplotlib import dates

font_path='../Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

DIR = '../Data_Jaesang/'
FILE = 'NaverSearchTrends_T.xlsx'
keyword=['날짜','발렌타인데이', '화이트데이']
```

```python
# 사용자 정의 함수
def typeChange(x):
    x['날짜']=x['날짜'].astype('datetime64')
    x.set_index('날짜',inplace=True)

def makeP(x):
    plt.figure(figsize=(10,7))
    plt.plot(x)

    plt.title('일별 검색량', fontsize=15)
    plt.xlabel('날짜', fontsize=12)
    plt.ylabel('검색 비율',fontsize=12)
    plt.legend(fontsize=12, labels=x.columns)
    plt.show()

def plotInfo(x):
    plt.title(f'{x}대 일일 검색량', fontsize=12)
    plt.xlabel('날짜', fontsize=12)
    plt.ylabel('검색 비율',fontsize=12)
    plt.xticks(rotation=40)
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    
def plotInfo2(x):
    plt.title(f'{x}대 일일 검색량', fontsize=12)
    plt.xlabel('날짜', fontsize=12)
    plt.ylabel('검색 비율',fontsize=12)
    plt.xticks(rotation=40)
    ax = plt.gca()
    ax.axes.yaxis.set_visible(False)
    
def makePlot(x1,x2,x3,x4,x5):
    plt.figure(figsize=(20,5))

    plt.subplot(151)
    plt.plot(x1)

    plt.title(gender, loc='left',fontsize=20)
    plt.title('10대 일일 검색량', fontsize=12)
    plt.xlabel('날짜', fontsize=12)
    plt.ylabel('검색 비율',fontsize=12)
    plt.legend(fontsize=12, labels=x1.columns, loc='best')
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)

    plt.subplot(152)
    plt.plot(x2)
    plotInfo(20)

    plt.subplot(153)
    plt.plot(x3)
    plotInfo(30)

    plt.subplot(154)
    plt.plot(x4)
    plotInfo(40)
    
    plt.subplot(155)
    plt.plot(x5)
    plotInfo(50)

    plt.show()

def makePlot2(x1,x2,x3,x4,x5):
    plt.figure(figsize=(20,5))

    plt.subplot(151)
    plt.plot(x1)

    plt.title(gender, loc='left',fontsize=20)
    plt.title('10대 일일 검색량', fontsize=12)
    plt.xlabel('날짜', fontsize=12)
    plt.ylabel('검색 비율',fontsize=12)
    plt.legend(fontsize=12, labels=x1.columns, loc='best')
    plt.xticks(rotation=40)

    plt.subplot(152)
    plt.plot(x2)
    plotInfo2(20)

    plt.subplot(153)
    plt.plot(x3)
    plotInfo2(30)

    plt.subplot(154)
    plt.plot(x4)
    plotInfo2(40)
    
    plt.subplot(155)
    plt.plot(x5)
    plotInfo2(50)

    plt.show()
```
